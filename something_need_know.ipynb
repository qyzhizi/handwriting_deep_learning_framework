## 图的构建与展示

```python
import networkx as nx
node_x, node_k1, node_b1 = 'x', 'k1', 'b1'
node_k2, node_b2 = 'k2', 'b2'
node_linear_01, node_linear_02, node_sigmoid = 'linear_01', 'linear_02', 'sigmoid'
node_loss = 'loss'

computing_graph = { # represent model 
   node_x: [node_linear_01],
   node_k1: [node_linear_01],
   node_b1: [node_linear_01],
   node_linear_01: [node_sigmoid],
   node_sigmoid: [node_linear_02],
   node_k2: [node_linear_02],
   node_b2: [node_linear_02],
   node_linear_02: [node_loss],
}
G = nx.Graph(computing_graph)
nx.draw(G, pos=nx.spring_layout(G), with_labels=True, node_color='red')
# plt.savefig("linear_simple.png")
```

```python
import networkx as nx
# 创建空的网格
G=nx.Graph()
# 添加节点
G.add_node('JFK')
G.add_nodes_from(['SFO','LAX','ATL','FLO','DFW','HNL'])
# G.number_of_nodes()  # 查看节点数，输出结果7

# 添加连线
G.add_edges_from([('JFK','SFO'),('JFK','LAX'),('LAX','ATL'),('FLO','ATL'),('ATL','JFK'),('FLO','JFK'),('DFW','HNL')])
G.add_edges_from([('OKC','DFW'),('OGG','DFW'),('OGG','LAX')])
# 绘制网络图
nx.draw(G,pos=nx.circular_layout(G),with_labels=True)

```

## 差集

在通用神经网络框架中的拓扑排序中出现过差集的运算，这里记录一下

a-b： 表示a对b的差集，得到一个新集合，里面的元素存在于a,但不存在于b。换句话说，如果a中的一个元素在集合b也存在，那么在集合a中删除这个元素，直到集合a中所有的元素都不存在于集合b。

```python
have_outputs = {5, 4, 2, 3}
have_inputs = {2,0,1,3}

only_outputs = have_outputs - have_inputs
print(only_outputs)
```

## numpy squeeze

https://numpy.org/doc/stable/reference/generated/numpy.squeeze.html

返回的东西：输入数组，但删除了长度为 1 的所有维度或维度的子集。这始终是 a 本身或 a 的视图。请注意，如果所有


```python
import numpy as np
x = np.array([[[0], [1], [2]]])
x.shape
(1, 3, 1)

np.squeeze(x).shape
```

```python
x_sq=np.squeeze(x)
print(x_sq)
```
